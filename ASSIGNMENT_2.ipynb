```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('electrical_dataset.csv')
```

```python
df[['Current (I)','Voltage (V)','Power (W)','Power_Factor','Temperature (°C)','Humidity (%)']] = df[['Current (I)','Voltage (V)','Power (W)','Power_Factor','Temperature (°C)','Humidity (%)']].fillna(df[['Current (I)','Voltage (V)','Power (W)','Power_Factor','Temperature (°C)','Humidity (%)']].mean())
```

```python
#Applying suitable methods to fill missing values in the dataset. For numerical columns usage of mean is more logical because mean minimizes error when predicting a value in a data set.
```

```python
df.head()
```

```python
#Detecting outliers in the 'Voltage (V)' and 'Power (W)' columns using the IQR method
```

```python
Q1 = np.percentile(df['Voltage (V)'], 25, method='midpoint')
Q3 = np.percentile(df['Voltage (V)'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)

upper = Q3+1.5*IQR
upper_array = np.array(df['Voltage (V)'] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())



lower = Q1-1.5*IQR
lower_array = np.array(df['Voltage (V)'] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
Q1 = np.percentile(df['Power (W)'], 25, method='midpoint')
Q3 = np.percentile(df['Power (W)'], 75, method='midpoint')
IQR = Q3 - Q1
print(IQR)

upper = Q3+1.5*IQR
upper_array = np.array(df['Power (W)'] >= upper)
print("Upper Bound:", upper)
print(upper_array.sum())


# Below Lower bound
lower = Q1-1.5*IQR
lower_array = np.array(df['Power (W)'] <= lower)
print("Lower Bound:", lower)
print(lower_array.sum())
```

```python
#Normalizing the 'Power (W)' column using Min-Max scaling
```

```python
min = df['Power (W)'].min()
max = df['Power (W)'].max()
```

```python
df['Power (W)'] = (df['Power (W)'] - min) / (max - min)
print(df['Power (W)'])
```

```python
#Assuming the Timestamp column represents data recorded during different shifts (e.g., Morning,
#Afternoon, Evening, Night). Creating a new categorical feature, ‘Shift’ . Grouping the dataset by Shift
#and calculating the average Power (W) and Temperature for each shift.
```

```python
df.loc[(df['Timestamp'].dt.time >= pd.to_datetime('05:00:00').time()) & (df['Timestamp'].dt.time <= pd.to_datetime('12:00:00').time()), 'Shift'] = 'Morning'
df.loc[(df['Timestamp'].dt.time >= pd.to_datetime('12:00:00').time()) & (df['Timestamp'].dt.time <= pd.to_datetime('17:00:00').time()), 'Shift'] = 'Afternoon'
df.loc[(df['Timestamp'].dt.time >= pd.to_datetime('17:00:00').time()) & (df['Timestamp'].dt.time <= pd.to_datetime('21:00:00').time()), 'Shift'] = 'Evening'
df.loc[((df['Timestamp'].dt.time >= pd.to_datetime('21:00:00').time()) & (df['Timestamp'].dt.time <= pd.to_datetime('23:59:59').time())) | ((df['Timestamp'].dt.time >= pd.to_datetime('00:00:00').time()) & (df['Timestamp'].dt.time <= pd.to_datetime('04:59:59').time())), 'Shift'] = 'Night'
```

```python
df.head()
```

```python
print(df.groupby('Shift')["Power (W)"].mean())
print(df.groupby('Shift')["Temperature (°C)"].mean())
```

```python
#Creating a new feature called 'Energy_Efficiency', calculating as 'Power (W) * Power_Factor'.
#Analysing the correlation between 'Energy_Efficiency' and 'Temperature (C)’.
```

```python
df['Energy_Efficiency'] = df['Power (W)']*df['Power_Factor']
```

```python
#Categorizing the 'Humidity (%)' into bins
```

```python
df.loc[(df['Humidity (%)'] >= 30) & (df['Humidity (%)'] <= 50), 'Humidity'] = 'Low'
df.loc[(df['Humidity (%)'] >= 50) & (df['Humidity (%)'] <= 60), 'Humidity'] = 'Medium'
df.loc[(df['Humidity (%)'] >= 60) & (df['Humidity (%)'] <= 70), 'Humidity'] = 'High'
```

```python
arr = df[['Voltage (V)','Current (I)','Humidity (%)','Energy_Efficiency']].to_numpy()
```

```python
arr
```

```python
#Split the data into train and test sets (80:20 ratio) after cleaning and preprocessing and creating the linear regression model
```

```python
x = arr
y = df['Power (W)'].to_numpy()

print("Features")
print(x)
print(x.shape)
print("Power (W)")
print(y)
print(y.shape)
#algorithm
l_reg = linear_model.LinearRegression()

x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2)

model = l_reg.fit(x_train, y_train)
predictions = model.predict(x_test)
print("Predictions: ",predictions)
print("R^2 value: ",l_reg.score(x, y))
print("Coeff:" , l_reg.coef_)
print("Intercept: ", l_reg.intercept_)

```
